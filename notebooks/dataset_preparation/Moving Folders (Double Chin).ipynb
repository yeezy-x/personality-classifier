```python
import os
import shutil
import random
from sklearn.model_selection import train_test_split

# ==========================================
# 1️⃣ Paths
# ==========================================

base_path = "/home/yeezy/100xDevs/personality/Personality-Classifier/"
source_base = os.path.join(base_path, "train_full_images")

train_base = os.path.join(base_path, "train_doublechin")
valid_base = os.path.join(base_path, "valid_doublechin")

classes = ["double_chin", "no_double_chin"]

# Create train & validation folders
for folder in [train_base, valid_base]:
    for cls in classes:
        os.makedirs(os.path.join(folder, cls), exist_ok=True)

# ==========================================
# 2️⃣ Split Each Class Separately
# ==========================================

for cls in classes:
    class_path = os.path.join(source_base, cls)
    images = os.listdir(class_path)

    train_imgs, valid_imgs = train_test_split(
        images,
        test_size=0.2,
        random_state=42
    )

    # Move train images
    for img in train_imgs:
        shutil.copy(
            os.path.join(class_path, img),
            os.path.join(train_base, cls, img)
        )

    # Move validation images
    for img in valid_imgs:
        shutil.copy(
            os.path.join(class_path, img),
            os.path.join(valid_base, cls, img)
        )

print("Train/Validation split complete.")
```
