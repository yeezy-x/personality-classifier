```python
import collections
import os
from collections import OrderedDict
FACIAL_LANDMARKS_IDXS = OrderedDict([
    ("mouth", (48, 68)),
    ('inner_mouth', (60, 68)),
    ("right_eyebrow", (17, 22)),
    ("left_eyebrow", (22, 27)),
    ("right_eye", (36, 42)),
    ("left_eye", (42, 48)),
    ("nose", (27, 35)),
    ("jaw", (0, 17))
])

print(os.listdir("/home/yeezy/100xDevs/personality/Personality-Classifier/ZipCeleb/img_align_celeba/img_align_celeba")[:5])
```

```python
# import the necessary packages
from imutils import face_utils
import numpy as np
import argparse
import imutils
import dlib
import cv2

# construct the argument parser and parse the argument
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

```

```python
DATA_DIR = "/home/yeezy/100xDevs/personality/Personality-Classifier/ZipCeleb/img_align_celeba/img_align_celeba"
REFINED_DIR = "/home/yeezy/100xDevs/personality/Personality-Classifier/facialImages/refined_images"
DISCARDED_DIR = "/home/yeezy/100xDevs/personality/Personality-Classifier/facialImages/discarded_images"

os.makedirs(REFINED_DIR, exist_ok=True) 
os.makedirs(DISCARDED_DIR, exist_ok=True) 

print("📂 Images found:", len(os.listdir(DATA_DIR)))

for img_name in os.listdir(DATA_DIR)[:10]:
    img_path = os.path.join(DATA_DIR, img_name)

    if not img_name.lower().endswith((".jpg", ".jpeg", ".png")):
        continue

    image = cv2.imread(img_path)
    if image is None:
        print("❌ Failed to read:", img_path)
        continue

    image = imutils.resize(image, width=750)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    rects = detector(gray, 1)
    stacked_coords = []

    for rect in rects:
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)

        for (name, (i, j)) in FACIAL_LANDMARKS_IDXS.items():
            for (x, y) in shape[i:j]:
                stacked_coords.append((x, y))

    is_valid_face = (len(rects) == 1 and len(stacked_coords) == 68)

    save_dir = REFINED_DIR if is_valid_face else DISCARDED_DIR
    save_path = os.path.join(save_dir, img_name)

    cv2.imwrite(save_path, image)
    print(f"{'✅ VALID' if is_valid_face else '❌ INVALID'} → {save_path}")

print("🎯 CelebA dataset filtering complete!")
```
