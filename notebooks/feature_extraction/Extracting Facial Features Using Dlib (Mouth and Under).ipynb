```python
import collections
from collections import OrderedDict
FACIAL_LANDMARKS_IDXS = OrderedDict([
    ("mouth", (48, 68)),
    ('inner_mouth', (60, 68)),
    ("right_eyebrow", (17, 22)),
    ("left_eyebrow", (22, 27)),
    ("right_eye", (36, 42)),
    ("left_eye", (42, 48)),
    ("nose", (27, 35)),
    ("jaw", (0, 17))
])
```

```python
from imutils import face_utils
import numpy as np
import argparse
import imutils
import dlib
import cv2
ap = argparse.ArgumentParser()
```

```python
detector = dlib.get_frontal_face_detector()
# Loads facial landmark predictor using path supplied
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

```python
image = cv2.imread('face.jpg') # takes from image headshot
# print(image)
if image is None:
    raise Exception("❌ Image not found. Check your image path!")

image = imutils.resize(image, width=600, height=600) # rescales to width of 500 px
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # converts to grayscale
# detect faces in the grayscale image
# Two arguments: grayscale image, number of image pyramid layers to apply when upscaling image prior to applying the detection
rects = detector(gray, 1) # creates the bounding box
if len(rects) == 0:
    raise Exception("❌ No face detected in the image!")

print(f"✅ Faces detected: {len(rects)}")
```

```python
# -----------------------------
# Process each detected face
# -----------------------------
for idx, rect in enumerate(rects):

    # Predict landmarks
    shape = predictor(gray, rect)
    shape = face_utils.shape_to_np(shape)

    # -----------------------------
    # Choose which facial part you want
    # Example: jaw
    # -----------------------------
    part_name = "jaw"
    (i, j) = FACIAL_LANDMARKS_IDXS[part_name]

    # Get bounding box around selected landmarks
    (x, y, w, h) = cv2.boundingRect(np.array([shape[i:j]]))

    # Add some padding below (useful for jaw)
    padding = 40
    y_start = max(0, y)
    y_end = min(image.shape[0], y + h + padding)
    x_start = max(0, x)
    x_end = min(image.shape[1], x + w)

    roi = image[y_start:y_end, x_start:x_end]

    # Resize ROI
    roi = imutils.resize(roi, width=250, inter=cv2.INTER_CUBIC)

    # -----------------------------
    # Save extracted feature
    # -----------------------------
    output_path = f"./extracted_{part_name}_{idx}.jpg"
    cv2.imwrite(output_path, roi)

    print(f"✅ Saved {part_name} ROI at: {output_path}")

    # -----------------------------
    # (Optional) Draw landmarks on image
    # -----------------------------
    clone = image.copy()
    for (x1, y1) in shape[i:j]:
        cv2.circle(clone, (x1, y1), 2, (0, 255, 0), -1)

    cv2.putText(clone, part_name, (10, 30),
    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

    # Show image (optional)
    # cv2.imshow("Landmarks", clone)
    # cv2.waitKey(0)

# cv2.destroyAllWindows()
```
