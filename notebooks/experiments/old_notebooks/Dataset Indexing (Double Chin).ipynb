```python
import pandas as pd
import os

base_path = "/home/yeezy/100xDevs/personality/Personality-Classifier/ZipCeleb/"
annotation_file = os.path.join(base_path, "list_attr_celeba.csv")

# ✅ Correct loading (comma separated)
data = pd.read_csv(annotation_file)

print("Columns:")
print(data.columns.tolist())

print("Total rows:", len(data))

# Extract only required columns
df = data[['image_id', 'Double_Chin']].copy()

# Rename for clarity
df.rename(columns={'image_id': 'Image_Name'}, inplace=True)

# Convert -1 → 0
df['Double_Chin'] = df['Double_Chin'].replace(-1, 0).astype(int)

print("\nClass Distribution:")
print(df['Double_Chin'].value_counts())

# Save CSV
output_path = "/home/yeezy/100xDevs/personality/Personality-Classifier/double_chin.csv"
df.to_csv(output_path, index=False)

print("\nDouble Chin indexing complete.")
print("Saved to:", output_path)
```
