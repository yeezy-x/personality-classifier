```python
import os
import shutil
import pandas as pd

base_path = "/home/yeezy/100xDevs/personality/Personality-Classifier/"
image_source = os.path.join(base_path, "ZipCeleb/img_align_celeba")
csv_path = os.path.join(base_path, "double_chin.csv")

output_base = os.path.join(base_path, "train_full_images")
double_chin = os.path.join(output_base, "double_chin")
no_double_chin = os.path.join(output_base, "no_double_chin")

os.makedirs(double_chin, exist_ok=True)
os.makedirs(no_double_chin, exist_ok=True)

df = pd.read_csv(csv_path)


for _, row in df.iterrows():
    image_name = row["Image_Name"]
    label = row["Double_Chin"]

    source_path = os.path.join(image_source, image_name)

    if not os.path.exists(source_path):
        continue

    if label == 1:
        destination = os.path.join(double_chin, image_name)
    else:
        destination = os.path.join(no_double_chin, image_name)

    shutil.copy(source_path, destination)

print("Images successfully organized into class folders.")
```
