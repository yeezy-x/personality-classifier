```python
import os
from os import path
import cv2
import dlib
import imutils
import numpy as np
import argparse
from imutils import face_utils
# construct the argument parser and parse the arguments
ap = argparse.ArgumentParser()
```

```python
BASE_PATH = os.path.expanduser("~/100xDevs/personality/Personality-Classifier")

INPUT_PATH = os.path.join(BASE_PATH, "data/raw/ZipCeleb")
OUTPUT_PATH = os.path.join(BASE_PATH, "data/processed/jaw")

PREDICTOR_PATH = os.path.join(BASE_PATH, "models/landmark_model/shape_predictor_68_face_landmarks.dat")

os.makedirs(OUTPUT_PATH, exist_ok=True)

print("Input:", INPUT_PATH)
print("Output:", OUTPUT_PATH)
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(PREDICTOR_PATH)
```

```python
count = 0

for file in os.listdir(INPUT_PATH):

    if not file.lower().endswith((".jpg", ".jpeg", ".png")):
        continue

    image_path = os.path.join(INPUT_PATH, file)
    image = cv2.imread(image_path)

    if image is None:
        continue

    image = imutils.resize(image, width=600)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    rects = detector(gray, 1)

    # Skip if not exactly one face
    if len(rects) != 1:
        continue

    for rect in rects:

        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)

        # Jaw region = 0 to 16
        start, end = 0, 17

        (x, y, w, h) = cv2.boundingRect(np.array([shape[start:end]]))
        roi = image[y:y + h + 50, x:x + w]

        try:
            roi = imutils.resize(roi, width=250)
            cv2.imwrite(os.path.join(OUTPUT_PATH, file), roi)
            count += 1
        except:
            continue

print("✅ Processing complete.")
print("Total images processed:", count)
```

```python
for (i, rect) in enumerate(rects):

    # determine facial landmarks
    shape = predictor(gray, rect)
    shape = face_utils.shape_to_np(shape)

    # Jaw region corresponds to landmark points 0–16
    start = 0
    end = 17

    print("Jaw region:", start, "to", end)

    # bounding box around jaw
    (x1, y1, w1, h1) = cv2.boundingRect(np.array([shape[start:end]]))

    # extract ROI
    roi = image[y1:y1 + h1 + 50, x1:x1 + w1]

    roi = imutils.resize(roi, width=250, inter=cv2.INTER_CUBIC)

    cv2.imwrite("headshot_jaw2.jpg", roi)
```

```python
clone = image.copy()

for (x, y) in shape[0:17]:
    cv2.circle(clone, (x, y), 2, (0, 255, 0), -1)

cv2.imshow("Jaw Landmarks", clone)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
