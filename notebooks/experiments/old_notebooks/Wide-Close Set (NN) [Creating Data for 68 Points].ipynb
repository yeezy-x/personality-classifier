```python

import collections
from collections import OrderedDict
from imutils import face_utils
import numpy as np
import argparse
import imutils
import dlib
import cv2

FACIAL_LANDMARKS_IDXS = OrderedDict([
    ("mouth", (48, 68)),
    ('inner_mouth', (60, 68)),
    ("right_eyebrow", (17, 22)),
    ("left_eyebrow", (22, 27)),
    ("right_eye", (36, 42)),
    ("left_eye", (42, 48)),
    ("nose", (27, 35)),
    ("jaw", (0, 17))
])


detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

```python
import os
import cv2
compiled_stacked_coors = []
for file1 in os.listdir('/content/gdrive/MyDrive/personality_datasets/train_full_images/big_nose'):
    print('Train Nose Big Nose', file1)
    image = cv2.imread('/content/gdrive/MyDrive/personality_datasets/train_full_images/big_nose/'+file1)
    # image = imutils.resize(image, width=1500, height=1500)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # detect faces in the grayscale image
    rects = detector(gray, 1)
    stacked_coors = []
    for (index, rect) in enumerate(rects):
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)

        a_dict = face_utils.FACIAL_LANDMARKS_IDXS.items()
        tuple_list = list(a_dict)
        # print(tuple_list)
        tuple_list = [tuple_list[6]]
        for f in range(len(tuple_list)):
            key_value = tuple_list[f]
            name = key_value[0]

            i = key_value[1][0]
            j = key_value[1][1]

            for (x, y) in shape[i:j]:
                stacked_coors.append((x,y))
  
    compiled_stacked_coors.append([stacked_coors, 1])

for file1 in os.listdir('/content/gdrive/MyDrive/personality_datasets/train_full_images/no_big_nose'):
    print('Train Nose No Big Nose', file1)
    image = cv2.imread('/content/gdrive/MyDrive/personality_datasets/train_full_images/no_big_nose/'+file1)
    # image = imutils.resize(image, width=1500, height=1500)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # detect faces in the grayscale image
    rects = detector(gray, 1)
    stacked_coors = []
    for (index, rect) in enumerate(rects):
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)

        a_dict = face_utils.FACIAL_LANDMARKS_IDXS.items()
        tuple_list = list(a_dict)
        # print(tuple_list)
        tuple_list = [tuple_list[6]]
        for f in range(len(tuple_list)):
            key_value = tuple_list[f]
            name = key_value[0]

            i = key_value[1][0]
            j = key_value[1][1]

            for (x, y) in shape[i:j]:
                stacked_coors.append((x,y))
  
    compiled_stacked_coors.append([stacked_coors, 0])

for file1 in os.listdir('/content/gdrive/MyDrive/personality_datasets/valid_full_images/big_nose'):
    print('Valid Nose Big Nose', file1)
    image = cv2.imread('/content/gdrive/MyDrive/personality_datasets/valid_full_images/big_nose/'+file1)
    # image = imutils.resize(image, width=1500, height=1500)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # detect faces in the grayscale image
    rects = detector(gray, 1)
    stacked_coors = []
    for (index, rect) in enumerate(rects):
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)

        a_dict = face_utils.FACIAL_LANDMARKS_IDXS.items()
        tuple_list = list(a_dict)
        # print(tuple_list)
        tuple_list = [tuple_list[6]]
        for f in range(len(tuple_list)):
            key_value = tuple_list[f]
            name = key_value[0]

            i = key_value[1][0]
            j = key_value[1][1]

            for (x, y) in shape[i:j]:
                stacked_coors.append((x,y))
  
    compiled_stacked_coors.append([stacked_coors, 1])

for file1 in os.listdir('/content/gdrive/MyDrive/personality_datasets/valid_full_images/no_big_nose'):
    print('Valid Nose No Big Nose', file1)
    image = cv2.imread('/content/gdrive/MyDrive/personality_datasets/valid_full_images/no_big_nose/'+file1)
    # image = imutils.resize(image, width=1500, height=1500)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # detect faces in the grayscale image
    rects = detector(gray, 1)
    stacked_coors = []
    for (index, rect) in enumerate(rects):
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)

        a_dict = face_utils.FACIAL_LANDMARKS_IDXS.items()
        tuple_list = list(a_dict)
        # print(tuple_list)
        tuple_list = [tuple_list[6]]
        for f in range(len(tuple_list)):
            key_value = tuple_list[f]
            name = key_value[0]

            i = key_value[1][0]
            j = key_value[1][1]

            for (x, y) in shape[i:j]:
                stacked_coors.append((x,y))
  
    compiled_stacked_coors.append([stacked_coors, 0])
```
